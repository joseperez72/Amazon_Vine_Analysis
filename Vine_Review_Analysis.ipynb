```python
import pandas as pd
```

```python
vine_table_df = pd.read_csv("vine_table.csv")
vine_table_df
```

```python
# create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20
step1_vine_df = vine_table_df.loc[vine_table_df['total_votes'] >= 20]
step1_vine_df
```

```python
# create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%
step2_vine_df = step1_vine_df.loc[(step1_vine_df['helpful_votes'] / step1_vine_df['total_votes']) >= 0.5]
step2_vine_df
```

```python
# create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid)
step3_vine_df = step2_vine_df.loc[step2_vine_df['vine'] == 'Y']
step3_vine_df.head()
```

```python
# create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (unpaid)
step4_vine_df = step2_vine_df.loc[step2_vine_df['vine'] == 'N']
step4_vine_df.head()
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
results = [
    {"vine_type": "Paid", 
     "total_review_count": len(step3_vine_df), 
     "5star_count": len(step3_vine_df.loc[step3_vine_df['star_rating'] == 5.0]), 
     "5star_percent": len(step3_vine_df.loc[step3_vine_df['star_rating'] == 5.0])/len(step3_vine_df) *100},
    {"vine_type": "Unpaid", 
     "total_review_count": len(step4_vine_df), 
     "5star_count": len(step4_vine_df.loc[step4_vine_df['star_rating'] == 5.0]), 
     "5star_percent": len(step4_vine_df.loc[step4_vine_df['star_rating'] == 5.0])/len(step4_vine_df) *100}
]
```

```python
results_df = pd.DataFrame(results)
results_df
```
